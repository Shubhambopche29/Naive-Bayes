```python
salary1 = salary.iloc[:,0:13]

salary1 = pd.get_dummies(salary1)
salary1
```

```python
finalsalary = pd.concat([salary1, salary['Salary']],axis=1)
finalsalary
```

```python
X = finalsalary.values[:,0:102]
Y = finalsalary.values[:,102]
```

```python
classifier_mb = MB()
classifier_mb.fit(X,Y)
train_pred_m = classifier_mb.predict(X)
accuracy_train_m = np.mean(train_pred_m==Y)

print('Training accuracy is:',accuracy_train_m*100,)
```
